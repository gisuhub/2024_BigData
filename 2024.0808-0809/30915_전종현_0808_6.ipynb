```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
sales = [1,5,7,3,7]
month = ['mar','apr','may','jun','jul']
#한글폰트 적용
plt.rc("font",family='Malgun Gothic')

#제목 
plt.title("월별 판매 실적")
plt.plot(month,sales,color='y',label="단위:개")
plt.xlabel("월")
plt.ylabel("매출")
plt.legend(loc = "lower right")
plt.show()
```
