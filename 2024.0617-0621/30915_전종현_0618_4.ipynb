```python
# 라이브러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt
# 한글폰트 적용하기
plt.rc('font',family='Malgun Gothic')
# 랜덤 라이브러리 불러오기
import random
# 데이터 만들기
height = []
weight = []
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
# 제목 : 키와 몸무게의 상관관계
plt.title("키와 몸무게의 상관관계")
# 산점도 그래프 그리기
plt.scatter(height,weight,c=height,cmap='RdPu')
# 축 제목
plt.colorbar(label='키')
plt.xlabel('키')
plt.ylabel("몸무게")
plt.show()
```

## 꺾은선 그래프 겹쳐 그리기

```python
import matplotlib.pyplot as plt

singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

plt.plot(singer,week1)
plt.plot(singer,week2)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')

singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

for i in range(len(week1)):
    week1[i] = -week1[i]
plt.rcParams["axes.unicode_minus"] =False

plt.title("오디션 프로그램 득표 현황")
plt.barh(singer,week1,color='pink',label='첫째주')
plt.barh(singer,week2,color='royalblue',label='둘째주')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')

singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

plt.title("오디션 프로그램 득표 현황")
plt.bar(range(1,6),week1,color='pink',label='첫째주')
plt.bar(range(7,12),week2,color='royalblue',label='둘째주')
plt.xticks(range(1,12),['A','B','C','D','E','','A','B','C','D','E'])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')

singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

plt.bar(range(1,14,3),week1,color='pink',label='첫째주')
plt.bar(range(2,15,3),week2,color='royalblue',label='둘째주')
plt.xticks(range(1,14,3),singer)
plt.legend()
plt.show()
```

```python
# 라이브러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt
# 한글폰트 적용하기
plt.rc('font',family='Malgun Gothic')
# 랜덤 라이브러리 불러오기
import random
# 데이터 만들기
height = []
weight = []
score = []
height1 = []
weight1 = []
score1 = []
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(0,200))
    height1.append(random.randint(100,200))
    weight1.append(random.randint(20,100))
    score1.append(random.randint(0,200))
# 제목 : 키와 몸무게의 상관관계
plt.title("키와 몸무게,체육 점수의 상관관계")
# 산점도 그래프 그리기
plt.scatter(height,weight,s=score,color='crimson',alpha=0.7,label='그룹1')
plt.scatter(height1,weight1,s=score1,color='indigo',alpha=0.7,label='그룹2')
# 축 제목
plt.legend()
plt.xlabel('키')
plt.ylabel("몸무게")
plt.show()
```
