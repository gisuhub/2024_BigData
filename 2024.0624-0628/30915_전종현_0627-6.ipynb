```python
import matplotlib.pyplot as plt

#한글폰트 적용하기
plt.rc("font",family='Malgun Gothic')

sales = [1,5,7,3,7]
month = ['mar','apr','may','jun','jul']

plt.figure(dpi=100)

plt.title("월별 판매 실적")
plt.plot(month,sales,color='m',label = '단위:개수')

plt.legend(loc='lower right')
plt.xlabel("월")
plt.ylabel("매출")

plt.show()
```

```python
# 라이브러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt
# 한글 폰트 불러오기
plt.rc("font",family='Malgun Gothic')
#랜덤 라이브러리 불러오기
import random
# 데이터 만들기
height,weight= [],[]
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    
# 제목 : 키와 몸무게의 상관관계
plt.title("키와 몸무게의 상관관계")
# 산점도 그래프 그리기
plt.scatter(height,weight,c=height,cmap='RdPu')
# 축 제목 
plt.colorbar(label='키')
plt.xlabel("키")
plt.ylabel("몸무게")

plt.show()
```

```python
# 라이브러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt
# 한글 폰트 불러오기
plt.rc("font",family='Malgun Gothic')
#랜덤 라이브러리 불러오기
import random
# 데이터 만들기
height = []
weight= []
score = []
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
# 제목 : 키와 몸무게의 상관관계
plt.title("키와 몸무게의 상관관계")
# 산점도 그래프 그리기
plt.scatter(height,weight,s=score,c=score,cmap='rainbow',alpha=0.7)

plt.colorbar(label='체육 실기 점수')

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')

singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

for i in range(len(week1)):
    week1[i] = -week1[i]
plt.rcParams["axes.unicode_minus"] =False

plt.title("오디션 프로그램 득표 현황")
plt.barh(singer,week1,color='hotpink',label='첫째주')
plt.barh(singer,week2,color='royalblue',label='둘째주')
plt.legend()
plt.show()
```
